```python
investors = {'GGV':'GGV纪源资本','IDG':'IDG资本','Ali':'阿里巴巴',
             'Baidu':'百度','HH_VC':'高瓴VC','HH':'高瓴资本','Sequoia':'红杉资本',
             'Matrix':'经纬中国','Qiming':'启明创投','Tencent':'腾讯投资','Zhen':'真格基金','BD':'字节跳动'}

list(investors.keys())
```

```python
def keyword(one_row):
    sentence = one_row['公司简介']
    lk = jieba.analyse.extract_tags(sentence,topK=5,allowPOS=('ns', 'n', 'vn', 'v'))
    one_row['关键词'] = '-'.join(lk)
    return one_row

def year_t(one_row):
    one_row['年份'] = one_row['时间'].year
    return one_row
```

```python
names = locals()
total_data = dict()
dates_year = [1998,1999,2000,2001,2002,2003,2004,2005,2006,2007, 2008,2009,2010,2011,2012,2013,2014, 2015,2016,2017,2018,2019,2020,2021]
total_data['year'] = dates_year
# col_names = ['year']
for i in list(investors.keys()):
    path = investors[i] + '.xlsx'
    names[i] = pd.read_excel(path)
    names[i]['关键词'] = '1'
    names[i] = names[i].apply(keyword,axis=1)
    names[i] = names[i][['时间','公司名','行业','子行业','轮次','金额','地址','关键词']]
    names[i]['时间']=pd.to_datetime(names[i]['时间'])
    names[i] = names[i].apply(year_t,axis=1)
#     dates.extend(list(names[i]['时间']))
    x=names[i]['年份'].value_counts()
    y = x.to_dict()
    num = [0]*len(dates_year)
    for year in dates_year:
        try:
            num_of_cyear = y[year]
        except:
            num_of_cyear = 0
        ind = dates_year.index(year)
        num[ind] = num_of_cyear
    total_data[investors[i]] = num
        

        
    
```
